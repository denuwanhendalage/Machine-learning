```python
#this is for ordinal encoding
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.DataFrame({'id':[1,2,3,4,5],'Education':['PhD','Masters','Bachelors','PhD','Masters'],'Country':['USA','UK','Canada','USA','USA']})
df
```

```python
df['Education'].unique()
```

```python
enc=OrdinalEncoder()
```

```python
enc
```

```python
enc.fit_transform(df[['Education']])
```

```python
df[['Education']]=enc.fit_transform(df[['Education']])
```

```python
education=['PhD','Masters','Bachelors']
```

```python
enc=OrdinalEncoder(categories=[education])
```

```python
df[['Education']]=enc.fit_transform(df[['Education']])
```
