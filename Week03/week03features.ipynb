```python
import pandas as pd
import numpy as np
```

```python
url='https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'
features=['preg','plas','pres','skin','test','mass','pedi','age','class']
df=pd.read_csv(url,names=features)
df.head()
```

```python
df.shape
```

```python
data=df.values
X=data[:,0:8]
Y=data[:,8]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

Filter methods

```python
# Feature extraction
chi_best=SelectKBest(score_func=chi2,k=4)
k_best=chi_best.fit(X,Y)

# Summarize scores
np.set_printoptions(precision=3)
print(k_best.scores_)

features = k_best.transform(X)

# Summarize selected features
print(features[0:5,:])
```

Wrapper method

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Feature extraction
model_lr= LogisticRegression()
recur_fe=RFE(model_lr,3)
Feature=recur_fe.fit(X,Y)
print("Number of Features: %s" % (Feature.n_features_))
print("Selected Features: %s" % (Feature.support_))
print("Feature Ranking is as follows : %s" % (Feature.ranking_))
```

Embedded Method

```python
from sklearn.linear_model import Ridge
```

```python
ridge_reg=Ridge(alpha=1.0)
ridge_reg.fit(X,Y)
```

```python
def print_coefs(coef,names=None,sort=False):
    if names==None:
        names=["X%s" % x for x in list(range(len(coef)))]

    lst=zip(coef,names)
    if sort:
        lst=sorted(lst,key=lambda x:-np.abs(x[0]))
    return "+".join("%s * %s" % (round(coef,names),name) for coef,name in lst)
```

```python
print("Ridge model:",print_coefs(ridge_reg.coef_))
```
